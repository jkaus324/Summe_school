```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/data-science-salaries-2023/ds_salaries.csv')
df.head()
```

```python
df.tail(10)
```

```python
df.dtypes
```

```python
pd.unique(df['job_title'])
```

```python
df = df.drop(['company_location'],axis=1)
df.head()
```

```python
df = df.rename(columns={"work_year":"year","experience_level":"experience","employment_type":"type","job_title":"title","salary":"sal","salary_currency":"currency","salary_in_usd":"usd","employee_residence":"residence","remote_ratio":"remote","company_size":"size"})
df.head()
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicated rows: ",duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x=df['sal'])
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['sal'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['title'],y=df['sal'])
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['title'],y=df['sal'])
```
